### API Calls

```python
url = '''https://api.census.gov/data/2020/dec/dhc'''
how = 'state:*'
where = 'county:*'

api_vars = [var.replace('\n', '').replace(' ', '') for var in api_vars]

dataframes = {}

for i, name in zip(api_vars, dataframe_names):
    r = requests.request("GET", url, params={
        'get': i,
        'for': where,
        'in': how,
        'key': key
    })

    try:
        json_data = r.json()
        dataframe = pd.DataFrame(json_data[1:], columns=json_data[0])
        dataframe['FIPS'] = dataframe['state'] + dataframe['county']
        del dataframe['state']
        del dataframe['county']
        dataframes[name] = dataframe
    except json.JSONDecodeError as e:
        print("JSON decoding error:", e)
        print("Response content:", r.text)

for name, cols in zip(dataframe_names, column_names):
    if name in dataframes:
        dataframe = dataframes[name]
        dataframe.columns = cols
        dataframes[name] = dataframe
    sleep(2)

globals().update(dataframes)
```
